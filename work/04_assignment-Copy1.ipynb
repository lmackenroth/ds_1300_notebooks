```python
sorted = countries.groupby('Tournament').mean()['Prize Money']


ax = sorted.plot.bar(x = 'Tournaments',y = 'Prize Money', legend = 'True' ,label='Prize Money Average', color = 'green', title = 'Countries vs Average prize money ',figsize=(10,10))
ax.set_xlabel("Countries")
ax.set_ylabel("Prize Money")
```
